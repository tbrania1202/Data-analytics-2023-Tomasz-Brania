Random data visualisation and sampling

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Exercise 2

```python
uniform_distribution = np.random.uniform(0, 1, 1000)
normal_distribution_0_1 = np.random.normal(0, 1, 1000)
normal_distribution_0_10 = np.random.normal(0, 10, 1000)
beta_distribution = np.random.beta(a=[1,2,3], b=[1,2,9], size=(1000, 3))

mean = np.array([0,0])
cov1 = np.array([[1,0], [0,1]])
cov2 = np.array([[2,1], [1,2]])

x1, y1 = np.random.multivariate_normal(mean, cov1, 10000).T
x2, y2 = np.random.multivariate_normal(mean, cov2, 10000).T
```

Exercise 3

```python
plt.figure(figsize=(10,8))
plt.suptitle('Histograms of one dimensional distributions')
plt.subplot(231)
plt.title('interval [0,1]')
plt.hist(uniform_distribution, bins=10)
plt.subplot(232)
plt.hist(normal_distribution_0_1, bins=10)
plt.title('normal(mean = 0, std = 1)')
plt.subplot(233)
plt.hist(normal_distribution_0_10, bins=10)
plt.title('normal(mean = 0, std = 10)')
plt.subplot(234)
plt.hist(beta_distribution[:,0], bins=10)
plt.title('beta distribution (1,1)')
plt.subplot(235)
plt.hist(beta_distribution[:,1], bins=10)
plt.title('beta distribution (2,2)')
plt.subplot(236)
plt.hist(beta_distribution[:,2], bins=10)
plt.title('beta distribution (3,9)')
plt.show()
```

Exercise 4

```python
plt.figure(figsize=(10,10))
plt.suptitle('Histograms (pyplot) for two-dimensionals distributions')
plt.subplot(321)
plt.hist2d(x1, y1)
plt.title('cov1 matrix, bin=default')
plt.subplot(322)
plt.hist2d(x2, y2)
plt.title('cov2 matrix, bin=default')
plt.subplot(323)
plt.hist2d(x1, y1, bins=20)
plt.title('cov1 matrix, bin=20')
plt.subplot(324)
plt.hist2d(x2, y2, bins=20)
plt.title('cov2 matrix, bin=20')
plt.subplot(325)
plt.hist2d(x1, y1, bins=50)
plt.title('cov1 matrix, bin=50')
plt.subplot(326)
plt.hist2d(x2, y2, bins=50)
plt.title('cov2 matrix, bin=50')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Histograms (seaborn) for two-dimensionals distributions')
plt.subplot(321)
sns.histplot(x=x1, y=y1)
plt.title('cov1 matrix, bin=default')
plt.subplot(322)
sns.histplot(x=x2, y=y2)
plt.title('cov2 matrix, bin=default')
plt.subplot(323)
sns.histplot(x=x1, y=y1, bins=20)
plt.title('cov1 matrix, bin=20')
plt.subplot(324)
sns.histplot(x=x2, y=y2, bins=20)
plt.title('cov2 matrix, bin=20')
plt.subplot(325)
sns.histplot(x=x1, y=y1, bins=100)
plt.title('cov1 matrix, bin=50')
plt.subplot(326)
sns.histplot(x=x2, y=y2, bins=100)
plt.title('cov2 matrix, bin=50')

plt.show()
```

Exercise 5

```python
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
std = abs(np.random.normal(0, 1, 1000))

u_param = alpha + beta*x

y = np.random.normal(u_param, std)
```

Exercise 6

```python
plt.figure()
plt.scatter(x, y)
plt.title('Normal distribution y')
plt.ylim([-8, 8])
plt.show()
```

```python
x_line = np.linspace(-6, 6, 1000)
y_line = []
for i in range(len(x_line)):
    y_line.append(alpha[i] + beta[i]*x[i])
```

```python

plt.figure()
plt.title('Normal distribution y')
plt.scatter(x, y)
for i in range(len(x_line)):
    plt.plot(x_line, x_line*y_line[i], c='orange', alpha=0.1)
plt.show()
```

```python
plt.figure()
plt.title('Normal distribution y')
plt.scatter(x, y)
for i in range(len(x_line)):
    plt.plot(x_line, x_line*y_line[i], c='orange', alpha=0.1)
plt.ylim([-8, 8])
plt.show()
```

Conclusions:

The histograms of normalized distributions in exercise 3 are different from the plots of analytical formulas for distribiutions. In principle, they have a similar shape, but they are not perfectly geometric, no matter how many bins are assigned. More bins affect more details on 2D histograms.
